```python
from scipy.stats import chi2_contingency
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data=[[50,142,131,70],
      [550,351,480,350]]
data
```

```python
pd.DataFrame(data,
             index = ['Males', 'Females'],
             columns = ['East','West','North','South'])
```

```python
plt.figure(figsize = (4,3))
names = ['East','West','North','South']

#Stacked Bar Plot
plt.bar(names, data[0])
plt.bar(names, data[1],bottom = data[0])


plt.xlabel('Region')
plt.ylabel('Genders')

plt.show()
```

```python
df = pd.DataFrame(data=[['Males',50,142,131,70],
                       ['Females',550,351,480,350]],
                  columns = ['','East','West','North','South'])

df
```

```python
df.plot(x = '',
        kind = 'bar',
        stacked = True)
```
